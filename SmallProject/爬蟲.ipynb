```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get('http://www.stat.nchu.edu.tw/member.php')
soup = BeautifulSoup(res.text)
output = []
for title in soup.select('div.teach_box'):
    print(title.text)
    data = title.select('h1')[0].text.strip().split('  ')
    for ele in title.select('p'):
        data.append(ele.text.split('  ', 1)[1])
    output.append(data)
```

```python
pd.DataFrame(output, columns = ["Name", "Title", "Lab", "Phone", "E-mail", "Domain"])
```
